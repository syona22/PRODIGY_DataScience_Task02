Data Cleaning

```python
#checking for null values
null_counts = df.isnull().sum()
```

```python
#bar chart for visual representation for null values found in each columns
null_counts.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Columns')
plt.ylabel('Number of Null Values')
plt.title('Null Values in Each Column')
plt.xticks(rotation=45)
plt.show()
```

```python
#count of null values in each columns
null_counts
```

```python
#number of null values for column Cabin are comparatively higher to include for further analysis so we discard the column
df = df.drop('Cabin', axis = 1)
df
```

```python
#filling the null values by median method for the column age. 
df['Age'].fillna(df['Age'].median(), inplace=True)
df
```

```python
#understanding the distribution of values and the common values for the column Embarked. 
df['Embarked'].value_counts()
```

```python
df['Embarked'].fillna('S', inplace=True)
df
```

```python
df.info()
```

```python
#changing the data types for the columns Survived, PClass, Sex, Age and Embarked 
df['Survived']=df['Survived'].astype('category')
df['Pclass']=df['Pclass'].astype('category')
df['Sex']=df['Sex'].astype('category')
df['Age']=df['Age'].astype('int')
df['Embarked']=df['Embarked'].astype('category')
df.info()
```

Exploratory Data Analysis

```python
df.describe()
```

```python
#Frequency distribution for Sex
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', data=df)
plt.title('Gender Distribution')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

```python
#Age distribution 
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['Age'].dropna(), kde=True, color = 'purple')
plt.title('Age Distribution')
```

```python
#Survival Analysis by Sex
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', hue='Sex', data=df, palette='Set1')
plt.title('Survival Count by Sex')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.legend(title='Sex', loc='upper right')
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'])  
plt.show()
```

```python
#Survival Analysis by PClass
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', hue='Pclass', data=df , palette = 'Set1')
plt.title('Survival Count by PClass')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.legend(title='PClass', loc='upper right')
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'])  
plt.show()
```

```python
#Distribution of Embarked
plt.figure(figsize=(8, 6))
sns.countplot(x = 'Embarked', data=df, palette = 'Set1')
plt.title('Embarked Distribution')
plt.xlabel('Port of Embarkation')
plt.ylabel('Count') 
plt.show()
```

```python
#scatter plot for Age and Fare
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Fare'], alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot: Age vs Fare')
plt.grid(True)
plt.show()
```

```python
correlation = df['Age'].corr(df['Fare'])
print("Correlation coefficient between Age and Fare:", correlation)
```

```python
#box plot for identifying outliers for the column Age
plt.figure(figsize=(8, 6))
sns.boxplot(x='Age', data=df)
plt.xlabel('Age')
plt.title('Box Plot: Distribution of Age')
plt.grid(True)
plt.show()
```

```python
#count plot for SibSp column
sns.countplot(x='SibSp', data=df)
plt.show()
```

```python
#count plot for Parch column
sns.countplot(x='Parch', data=df, palette = 'Set1')
plt.show()
```

```python
#distribution of passengers based on family size to know who survived. 
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
plt.figure(figsize=(10, 6))
sns.countplot(x='FamilySize', hue='Survived', data=df, palette='Set1')
plt.title('Survival Count by Family Size')
plt.xlabel('Family Size')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```
